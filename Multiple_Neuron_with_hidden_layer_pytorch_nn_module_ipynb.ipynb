```python
# Import necessary libraries
import torch
import torch.nn as nn

# Define a neural network model class
class Model(nn.Module):
    def __init__(self, num_features):
        # Call the parent class (nn.Module) constructor
        super().__init__()

        # Define a linear layer with 'num_features' input and 1 output
        self.linear1 = nn.Linear(num_features, 3)
        self.relu = nn.ReLU()
        self.linaer2 = nn.Linear(3,1)
        # Define a sigmoid activation function
        self.sigmoid = nn.Sigmoid()

    # Define the forward pass
    def forward(self, features):
        # Apply the linear layer to the input features
        out = self.linear1(features)
        out = self.relu(out)       #---# Hidden Layer ma RAelu
        out = self.linaer2(out)
        # Apply the sigmoid activation function
        out = self.sigmoid(out)     # --- Output - Sigmoid

        # Return the output
        return out
```

```python
# Create Datasets
features = torch.rand(10,5)

# Create Model
model = Model(features.shape[1])

# Call model for forward pass
# model.forward(features) --optional
# Correct way

model(features)
```
